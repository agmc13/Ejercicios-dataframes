```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class DataFrame:
    global prueba
    prueba= []
    def __init__(self):
        global df 
        
        if True:
            try:
                df= pd.read_excel('NBA.xlsx', sheet_name=0, header=0, index_col=False, keep_default_na=True)
                print (df)
            except ValueError:
                print('no se ha podido crear el dataframe correctamente')
    def grafica(self,posicion,valor2):
        global prueba
        if True:
            try:
                zona= df.groupby(['Posicion']).get_group((posicion))
                zona.groupby([valor2,'Nombre'], sort=False)['Salario'].max().unstack(fill_value=0).plot(kind='bar',stacked=False)               
                maximo= zona[zona[valor2]==zona[valor2].max()]
                prueba.extend(maximo.values)
                plt.title(posicion)
                plt.show()
            except ValueError:
                print('no se ha podido crear la grafica correctamente')
    def total(self):
        if True:
            try:
                df2 = pd.DataFrame(prueba, columns=('Nombre','Equipo','Estatura','Salario','Posicion','Rebotes','Robos','Triples','Asistencia','Puntos','Edad','Puntuacion'))
                df2.groupby(['Posicion','Nombre'], sort=False)['Salario'].max().unstack(fill_value=0).plot(kind='bar',stacked=False)
                plt.title('Equipo que vamos a tener')
                plt.show()
            except ValueError:
                print('no se ha podido crear la grafica correctamente')      
```

```python
dato = DataFrame()
```
